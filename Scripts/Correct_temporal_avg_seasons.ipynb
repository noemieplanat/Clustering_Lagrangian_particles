```python
import xarray as xr
import numpy as np
import glob
```

```python
files = glob.glob('/storage/nplanat/Glorys12_ORCA_seasons/'+'*.nc')
```

```python
for f in files:
    print(f[-23:-4])
    ds = xr.open_dataset(f)
    ds['x'] = ('x', np.arange(1, ds.nav_lat.shape[1]+1))
    ds['y'] = ('y', np.arange(1, ds.nav_lat.shape[0]+1))
    ds = ds.set_coords({'x', 'y'})
    ds.to_netcdf('/storage/nplanat/Glorys12_ORCA_seasons/'+'NEW_'+f[-23:-4]+'.nc')
```
